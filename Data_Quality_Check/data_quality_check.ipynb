```python
import numpy as np
import pandas as pd
import seaborn as sns
```

Load Dataset

```python
# Load the built-in Titanic dataset
df = sns.load_dataset('titanic')

# Look at the first 5 rows to understand the columns
df.head()
```

Adding few columns for demonstrating data issues

```python
df.head()
```

```python
# Add a new column 'passenger_id' with unique values for each passenger
df['passenger_id'] = np.arange(1, len(df)+1)

# Add a constant column with the same value for all rows
df['contant_col'] = 1

#create a dirty version of embark_town column
df['embark_town_dirty'] = df['embark_town'].copy()

# select a few random rows only non-missing values
sample_idx = df['embark_town_dirty'].dropna().sample(6, random_state=42).index

# Apply simple dirty transformations to the selected rows
df.loc[sample_idx [0:2], 'embark_town_dirty'] = df.loc[sample_idx [0:2], 'embark_town_dirty'].str.upper()  # Convert to uppercase
df.loc[sample_idx [2:4], 'embark_town_dirty'] = df.loc[sample_idx [2:4], 'embark_town_dirty'].str.strip().apply(lambda x: f' {x} ')  # Add extra spaces
df.loc[sample_idx [4:6], 'embark_town_dirty'] = 'unknown'  # Replace with unknown

df.head()
```
